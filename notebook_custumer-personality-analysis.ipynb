### 3.3 Daten transformieren <a class="anchor" id="3.3 Daten transformieren"></a>

Checkliste für den Abschnitt "Datentransformation": 

a) Feature engineering:
- [ ] Extrahieren des "Alters" eines Kunden mit Hilfe von "Year_Birth", welches das Geburtsjahr der jeweiligen Person angibt.
- [ ] Erstellen eines Merkmals "Spent", das den Gesamtbetrag angibt, den der Kunde in verschiedenen Kategorien über einen Zeitraum von zwei Jahren ausgegeben hat.
- [ ] Erstellen eines Merkmals "Living_With" aus "Marital_Status", um die Lebenssituation von Paaren zu ermitteln.
- [ ] Erstellen eines Merkmals "Kinder", um die Gesamtzahl der Kinder in einem Haushalt anzugeben, d. h. Kinder und Teenager.
- [ ] Erstellen eines Merkmals "Family_Size", um mehr Klarheit über den Haushalt zu erhalten.
- [ ] Erstellen eines Merkmals "Ist_Elternteil", um den Status der Elternschaft zu berücksichtigen.
- [ ] Bildung von drei Kategorien im Bereich "Bildung", um die Werte zu vereinfachen.
- [ ] Zuguterletzt Weglassen einiger überflüssiger Merkmale zur Reduktion der Komplexität.

b) Data Pre-processing:
- [ ] Labelencoding
- [ ] StandardScaling

c) Dimensionality reduction:
- [ ] optional PCA

```python
# Hinzufügen einer neuen Spalte mit dem aktuellen Alter (ausgehend vom Geburtsjahr)
marketing["Age"] = 2015 - marketing["Year_Birth"]
```

```python
# Vereinheitlichung der Merkmale zum Marital_Status
marketing["Marital_Status"] = marketing["Marital_Status"].replace(
    ["Married", "Together"], "Taken")
marketing["Marital_Status"] = marketing["Marital_Status"].replace(
    ["Divorced", "Widow", "Alone", "Absurd", "YOLO"], "Single")
```

```python
marketing['Marital_Status'].value_counts()
```

```python
# Vereinheitlich der Merkmale zu Kids und Teenagern zu Minderjährige (Minors)
marketing["Minors"] = marketing["Kidhome"] + marketing["Teenhome"]
```

```python
# Erstellen eines neuen Merkmals names "Has_child"
marketing['Has_child'] = np.where(marketing.Minors > 0, 'Yes', 'No')
```

```python
# Zusammenfassung der Ausgaben für die einzelnen Produktkategorien zu Ausgaben_Gesamt (Expenses)
marketing["Expenses"] = marketing["MntWines"] + marketing["MntFruits"] + marketing["MntMeatProducts"] + \
    marketing["MntFishProducts"] + \
    marketing["MntSweetProducts"] + marketing["MntGoldProds"]
```

```python
# Zusammenfassung der akzetierten Marketingkampagnen (1-5) zu einem Merkmal (AcceptedCmp_Total)
marketing["AcceptedCmp_Total"] = marketing["AcceptedCmp1"] + marketing["AcceptedCmp2"] + \
    marketing["AcceptedCmp3"] + marketing["AcceptedCmp4"] + \
    marketing["AcceptedCmp5"] + marketing["Response"]
```

```python
# Zusammenfassung der Anzahl der Käufe über unterschiedliche Kanäle (Online, Geschäft, Katalog) zu einem Merkmal (NumPurchases_Total)
marketing["NumPurchases_Total"] = marketing["NumWebPurchases"] + \
    marketing["NumCatalogPurchases"] + \
    marketing["NumStorePurchases"] + marketing["NumDealsPurchases"]
```

## 4. Modelling <a class="anchor" id="4. Modelling"></a>

### 4.1 Modellauswahl <a class="anchor" id="4.1 Modellauswahl"></a>

In dieser Projektarbeit wird eine# unsupervised learning Model verwerdet. Anders als beim supervised learning (z.B. Klassifikations- oder Regressionmodelle) besteht im Datensatz kein Marker bzw. keine markierten Ergebnisse, welche bereits die richtigen Resultate für Zusammenhänge im Datensatz darstellen und so zum trainieren eines Machine Learning Modells verwendet werden können. 

Als Beispiel könnte ein Datensatz basierend auf gemessenen Merkmalen zu sportlichen Aktivitäten wie Herzfrequenz, Körpergewicht, Körpergröße, Alter, Dauer der Aktivität und dem zugehörigen Kalorienverbrauch dazu genutzt werden, den Kalorienverbrauch vorherzusagen. Dazu werden im sogenannten Train-Test-Split ein Trainings- und ein Testdatensatz aus den Ursprungsdaten gebildet. Ein Regressionsmodell würde dann mit den Trainingsdaten trainiert. Anschließend können die Testdaten zur Bestimmung der Genauigkeit des Modells herangezogen werden. Ein Regressionmodell könnte nun anhand von zufälligen Angaben zu Herzfrequenz, Körpergewicht, Körpergröße, Alter, Dauer der Aktivität den voraussichtlichen Kalorienverbrauch berechnen.

Beim unsupervised learning liegen diese hilfreichen Rsultate vorab nicht vor. Vielmehr werden z.B. "Muster" oder Zusammenhänge in einem Datensatz identifiziert. Ein Train-Test-Split findet beim unsupervised learning ebenfalls nicht statt. Zusammengefasst lässt sich sagen, im Gegensatz zum überwachten Lernen wird das unüberwachte Lernen verwendet, um Rückschlüsse zu ziehen und Muster aus den Eingabedaten zu finden, ohne auf markierte Ergebnisse Bezug zu nehmen.

Das gängigste und spannendste Modell im Bereich unüberwachtes Lernen/Clustering für dieses Projekt ist das: `K-Means Clustering`. Das K-Means Clustering ist eine Methode des unüberwachten Lernen und versucht gleiche Daten zusammenzufassen und unterschiedliche Daten zu separieren. Dabei werden die Abstände gleichartiger Daten innerhalb eines Clusters möglichst minimiert und der Abstand gegenüber anderen Clustern möglichst maximiert. Ziel ist es dabei, eine passende Anzahl von k Clustern zu finden, so dass die partitionierten Datenpunkte innerhalb eines Cluster möglichst "ähnlich" sind. Die Ähnlichkeit zweier Punkte zeigt sich durch deren Abstand zueinander. Dieser kann auf unterschiedliche Arten gemessen werden. Der euklidische Abstand ist dabei eine gängige Messvariante.

Weitere interessante Ansätze, jedoch hier nicht weiter im Detail betrachtet, sind das hierarchische Clustering, das Mean-Shift-Clustering und das Density Based Clustering.

### 4.2 Modell aufsetzen <a class="anchor" id="4.2 Modell aufsetzen"></a>

Wie bereits in der Modellauswahl beschrieben, ist es wichtig, die richtige Anzahl von k Clustern so zu bestimmen, dass die Segmentierung möglichst sinnvoll in die Cluster übertragbar ist. Dabei hilft und die sogennante `Elbow Method`.

**Elbow Method**

Jedes Mal wenn wir ein weiteres Cluster hinzufügen, also K + 1, verringert sich die Varianz. Die Reduktion der Varianz wird immer kleiner. Wenn wir K = N hätten, wobei N für die Anzahl der Datenpunkte steht, wäre die Varianz = 0. Aber dann brauchen wir auch keine Cluster mehr. Der Punkt, ab dem die zusätzliche erklärte Varianz bei Zunahme eines weiteren Clusters stark sinkt, nennt sich „elbow“. Deswegen wird die Darstellung auch “elbow plot“ genannt.

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(9, 5))
plt.plot(range(1, 11), wcss, 'bx-')

plt.title('Elbow plot')
plt.xlabel('Anzahl an Cluster')
plt.ylabel('WCSS')
```

```python
# Schnelle Darstellung zur Findung der richtigen Clusteranzahl mit Hilfe der Elbow Methode
print('Elbow Method:')
Elbow_M = KElbowVisualizer(KMeans(), k=10)
Elbow_M.fit(X)
Elbow_M.show()
```

Mit Hilfe der Elbow-Method und dem KElbowVisualizer wird die passende Anzahl an Cluster für den vorliegenden Dataframe mit 4 festgelegt.

Die Einteilung der Kunden in die Cluster basiert dabei auf vielen unterschiedlichen Merkmalen, wie die Familienstruktur, dem Einkommen, den Ausgaben sowie dem gerellen Kaufverhalten. Diese Kundensegmentierung kann zukünftig als Basis für die verbesserte Planung unserer Marketingstrategien genutzt werden. Einige Ansätze und Vorschläge wurden mit dieser Arbeit gemacht.
